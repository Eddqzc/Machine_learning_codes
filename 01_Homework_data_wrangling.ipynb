## Excersise 1 NFL
Given a dataset of NFL players containing: Age, Birth Place, Birthday, College, Current Status, Current Team, Experience, Height (inches), High School, High School Location, Name, Number, Player Id, Position, Weight (lbs), Years Played; In this dataset there are a lot os empty cell.

    Filter out the NaN values and fill up:
    With the mean of the column
    With random values between the min and the max of the column

```python
nfl = pd.read_csv('data/NFL.csv',parse_dates=['Birthday'])
nfl
```

```python
nfl.info()
```

### Nan treatments

```python
db = nfl.copy()
db['Height (inches)']= np.where(db['Height (inches)'].isnull(),
                                random.randrange(db['Height (inches)'].min(),db['Height (inches)'].max()),
                                db['Height (inches)'])
db['Number']= np.where(db['Number'].isnull(),
                       random.randint(db['Number'].min(),db['Number'].max()),db['Number'])
db['Player Id']= np.where(db['Player Id'].isnull(),
                          db['Name'].str.replace(', ','').str.lower()+'/0000000',db['Player Id'])
db['Weight (lbs)']= np.where(db['Weight (lbs)'].isnull(),
                             random.randrange(db['Weight (lbs)'].min(),db['Weight (lbs)'].max()),
                             db['Weight (lbs)'])

values_fill = {'Birth Place':db['Birth Place'].mode()[0],
               'Birthday':db['Birthday'].mean(),
               'College':db['College'].mode()[0],
               'Current Status':'Retired',
               'Current Team':'No team',
               'Experience':db['Experience'].mode()[0],
               #'Height (inches)':db['Height (inches)'].mean(),
               'High School':db['High School'].mode()[0],
               'High School Location':db['High School Location'].mode()[0],
               'Name':'no name',
               #'Number':int(db['Number'].mean()),
               #'Player Id':db['Name'].str.replace(', ','').str.lower()+'/0000000',
               'Position':db['Position'].mode()[0],
               #'Weight (lbs)':db['Weight (lbs)'].mean(),
               'Years Played':db['Years Played'].mode()[0],
              }
db.fillna(value=values_fill,inplace = True)
db['Age'] = datetime.now().year - db['Birthday'].dt.year
db
```

```python
data = pd.read_csv('data/Flights/nycflights13.csv')
data
```

### Rename columns

```python
# Rename Columns
db = data.copy()
cols={'dep_time':'deperture_time','dep_delay':'deperture_delay','arr_time':'arrive_time',
      'arr_delay':'arrive_delay','dest':'destiny'}
db.rename(cols,axis=1,inplace=True)
db
```

### Filter flights

```python
# From or to JFK
jfk = db[(db['origin']=='JFK')|(db['destiny']=='JFK')]
jfk
```

### Sorting data

```python
jfk['aux'] = np.where(jfk['origin']=='JFK',1,2)
jfk = jfk.sort_values(['aux','arrive_delay'])
jfk.reset_index(drop=True,inplace=True)
jfk
```
